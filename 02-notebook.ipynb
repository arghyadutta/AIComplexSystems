# Few points on AI

**Artificial Intelligence (AI)**: Simulation of human intelligence in machines that are programmed to think, reason, and learn. 

Types:

1. **Narrow AI**: Perform specific tasks (basically ML)
2. **General AI**: *Hypothetical* systems capable of performing any intellectual task a human can do.
3. **Superintelligent AI**: Surpassing human intelligence. *Hypothetical* again.

**Ethics: Many—many— unresolved issues!**

- Data Privacy
- Algorithmic bias
- People losing/not getting jobs
- Intellectual property rights
- Danger from superintelligent AI

**Uses**

- Do homework assignments with ChatGPT?
- Recommender systems (Netflix)
- Translation (DeepL)
- Speech understanding (Siri)...

**Learning algorithms** are computational methods used by machines to improve their performance on tasks over time based on experience (data).

**Supervised Learning**:

Learning from labeled data. Examples: Regression, Classification. Algorithms: Linear Regression, Decision Trees, Neural Networks.

### Not accurate, but not very wrong either; at least for the narrow AI.

![image.png](./images/stat-ml-meme.png)

**Unsupervised Learning**:

Learning from unlabeled data. Clustering, Dimensionality Reduction (K-Means, PCA)

**Reinforcement Learning**
   
Learning through interaction with the environment.Examples: Game-playing AI (Q-Learning)


## A quick example

![](images/iris.png)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
iris_dataset = load_iris()
type(iris_dataset)

for key in iris_dataset.keys():
    print(key + "\n", iris_dataset[key])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset["data"], iris_dataset["target"]
)

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
iris_dataframe
```

```python
# create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(
    iris_dataframe, c=y_train, marker="o", s=20, figsize=(10, 10)
)
pass;
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
X_new = np.array([[5, 2.9, 1, 0.2]])

prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Predicted target name:", iris_dataset["target_names"][prediction])
pass
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```
